You can use SymPy to solve a simple equation:

```python
from sympy import symbols, Eq, solve

# Define the variable
equation = Eq(2*x + 5, 11)

# Solve the equation
solution = solve(equation, x)

# Print the solution
print("Solution:", solution)
```

SymPy can be used to solve simultaneous equations. Here's an example:

```python
# Define the variables
x, y = symbols('x y')

# Define the system of equations
equation1 = Eq(2*x + y, 5)
equation2 = Eq(x - 2*y, 15)

# Define the system of equations
system_of_equations = [equation1, equation2]

# Solve the system of equations
solution = solve(system_of_equations, (x, y))

# Display the solution
print("Solution:", solution)
```

Limit Computation

Suppose that you want to find the limit of a function as it approaches a certain value. Here's an example using SymPy to do so:

```python
from sympy import limit, symbols
from sympy import sin

# Define the function
f = sin(x) / x

# calculate the limit as x approaches 0
lim_result = limit(f, x, 0)
lim_result
```

Derivative Computation

```python
from sympy import diff
from sympy import sin

# Define the variable 
x = symbols('x')

# Define the function
f = x**3 + 3 * x**2 + sin(x)

# Calculate the derivative
def_f = diff(f, x)

def_f
```

Integral Computation

SymPy can be used to compute both definite and indefinite integrals.

```python
from sympy import symbols, integrate, sin

# Define the variable
x = symbols('x')

# Define the function
f = x*sin(x)

# Compute the indefinite integral
indefinite_integral = integrate(f,x)
indefinite_integral 
```

If you want to evaluate it, you can compute the derivative of the result, and you must get the initial function xsin(x).
One can also compute the definite integral using SymPy. Here's an example:

```python
from sympy import cos, pi

# Define the variable
x = symbols('x')

# Define the function
f = cos(x)

# Compute the definite integral
definite_integral = integrate(f, (x,0,pi/2)) 
definite_integral
```

Taylor Series

Here's an example of computing the Taylor series for e^x around x=0 up to the 4th degree:

```python
from sympy import symbols, exp
# Define the variable and the function
x = symbols('x')
f = exp(x)

# Compute the terms of the Taylor series
taylor_series = f.series(x, 0, 4).removeO()

# Display the terms of the Taylor series
print(taylor_series)
```

Least Squares Problem

```python
import sympy as sp

# Sample data points
data_points = [(1,2), (2,3), (3,4), (4,5)]

# Variables for the linear equation: y = mx + c
m, c = sp.symbols('m c')

# Sum of squared differences between observed and predicted y-values
error = sum((m * x + c - y)**2 for x, y in data_points)

# Finding partial derivatives of the error function with respect to m and c
partial_m = sp.diff(error, m)
partial_c = sp.diff(error, c)

# Solving the system of equations to minimise the error (least squares solution)
solution = sp.solve((partial_m, partial_c), (m, c))

best_fit_m, best_fit_c = solution[m], solution[c]
```

Now, use the least squares method to find the parameters for the following models:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

# Model 1: y = a1x + a0
coefficients_model1 = np.polyfit(x, y, deg=1)
y_model1 = np.polyval(coefficients_model1, x)

# Model 2: y = a2x + a1x + a0
coefficients_model2 = np.polyfit(x, y, deg=2)
y_model2 = np.polyval(coefficients_model2, x)

# Model 3: y = a3x + a2x + a1x + a0
coefficients_model3 = np.polyfit(x, y, deg=3)
y_model3 = np.polyval(coefficients_model3, x)

# Model 4: y = a4x + a3x + a2x + a1x + a0
coefficients_model4 = np.polyfit(x, y, deg=4)
y_model4 = np.polyval(coefficients_model4, x)

# Model 5: y = a5x + a4x + a3x + a2x + a1x + a0
coefficients_model5 = np.polyfit(x, y, deg=5)
y_model5 = np.polyval(coefficients_model5, x)

# Plotting
plt.plot(x, y_true, '--', label='True Function')
plt.plot(x, y, '.', label='Noisy Data')
plt.plot(x, y_model1, label='Model 1')
plt.plot(x, y_model2, label='Model 2')
plt.plot(x, y_model3, label='Model 3')
plt.plot(x, y_model4, label='Model 4')
plt.plot(x, y_model5, label='Model 5')
plt.xlabel('x', size=14)
plt.ylabel('y', size=14)
plt.legend()
plt.show()

# Display coefficients
print('Model 1 Coefficients:', coefficients_model1)
print('Model 2 Coefficients:', coefficients_model2)
print('Model 3 Coefficients:', coefficients_model3)
print('Model 4 Coefficients:', coefficients_model4)
print('Model 5 Coefficients:', coefficients_model5)


# Calculate predictions for each model
y_pred1 = np.polyval(coefficients_model1, x)
y_pred2 = np.polyval(coefficients_model2, x)
y_pred3 = np.polyval(coefficients_model3, x)
y_pred4 = np.polyval(coefficients_model4, x)
y_pred5 = np.polyval(coefficients_model5, x)

# Calculate mean squared error for each model
mse1 = mean_squared_error(y, y_pred1)
mse2 = mean_squared_error(y, y_pred2)
mse3 = mean_squared_error(y, y_pred3)
mse4 = mean_squared_error(y, y_pred4)
mse5 = mean_squared_error(y, y_pred5)

# Print mean squared error for each model
print('Mean Squared Error (Model 1):', mse1)
print('Mean Squared Error (Model 2):', mse2)
print('Mean Squared Error (Model 3):', mse3)
print('Mean Squared Error (Model 4):', mse4)
print('Mean Squared Error (Model 5):', mse5)
```

```python
from sympy import symbols, diff, sqrt, cot, acos, sin, cos, ln, exp, atan, log

x = symbols('x')

# 1. Derivative of y = x**2 + 2*x + 1
y = x**2 + 2*x + 1
dy_dx = diff(y, x)
print('Derivative of 1 is:   ',dy_dx)

#2. Derivative of y = (3*x - 5)**3
y = (3*x - 5)**3
dy_dx = diff(y, x)
print('Derivative of 2 is:   ',dy_dx)

#3. Derivative of y = (sqrt(x) - 1)**2 - (x**2 + 1)**4
y = (sqrt(x) - 1)**2 - (x**2 + 1)**4
dy_dx = diff(y, x)
print('Derivative of 3 is:   ',dy_dx)

#4. Derivative of y = 7*cot(x) - 8*acos(x)
y = 7*cot(x) - 8*acos(x)
dy_dx = diff(y, x)
print('Derivative of 4 is:   ',dy_dx)

#5. Derivative of y = x - ln(x) + 7
y = x - ln(x) + 7
dy_dx = diff(y, x)
print('Derivative of 5 is:   ',dy_dx)

#6. Derivative of y = -10*exp(x) + 5**x - 5*x + x/5
y = -10*exp(x) + 5**x - 5*x + x/5
dy_dx = diff(y, x)
print('Derivative of 6 is:   ',dy_dx)

#7. Derivative of y = 2*sin(x)/(sin(x) - cos(x))
y = 2*sin(x)/(sin(x) - cos(x))
dy_dx = diff(y, x)
print('Derivative of 7 is:   ',dy_dx)

#8. Derivative of y = (x**2 * log(x))/(1 - atan(x))
y = (x**2 * log(x))/(1 - atan(x))
dy_dx = diff(y, x)
print('Derivative of 8 is:   ',dy_dx)
```

<span style='font-size: larger;'>**TASK 5** - Find the Taylor series of degree 5 for the following functions at the specified points. Utilise Python to plot both the functions and their respective Taylor series, then compare the two.

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the symbolic variable
x = sp.symbols('x')

# Define the functions
f1 = sp.sin(x)
f2 = sp.exp(3*x)
f3 = sp.ln(x)
f4 = sp.tan(x)

# Specify the points
point1 = 0
point2 = 0
point3 = 1
point4 = sp.pi/4

# Compute the Taylor series of degree 5
taylor_series1 = sp.series(f1, x, point1, 6).removeO()
taylor_series2 = sp.series(f2, x, point2, 6).removeO()
taylor_series3 = sp.series(f3, x, point3, 6).removeO()
taylor_series4 = sp.series(f4, x, point4, 6).removeO()

# Convert the symbolic expressions to Python functions
taylor_func1 = sp.lambdify(x, taylor_series1, 'numpy')
taylor_func2 = sp.lambdify(x, taylor_series2, 'numpy')
taylor_func3 = sp.lambdify(x, taylor_series3, 'numpy')
taylor_func4 = sp.lambdify(x, taylor_series4, 'numpy')

# Create a range of x values for plotting
x_values = np.linspace(-2, 2, 400)

# Evaluate the functions and their Taylor series at the x values
y_values1 = np.sin(x_values)
y_taylor1 = taylor_func1(x_values)

y_values2 = np.exp(3*x_values)
y_taylor2 = taylor_func2(x_values)

y_values3 = np.log(x_values)
y_taylor3 = taylor_func3(x_values)

y_values4 = np.tan(x_values)
y_taylor4 = taylor_func4(x_values)

# Plot the functions and their Taylor series
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x_values, y_values1, label='sin(x)')
plt.plot(x_values, y_taylor1, label='Taylor Series (degree 5)')
plt.title('sin(x) at x=0')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, y_values2, label='e^(3x)')
plt.plot(x_values, y_taylor2, label='Taylor Series (degree 5)')
plt.title('e^(3x) at x=0')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x_values, y_values3, label='ln(x)')
plt.plot(x_values, y_taylor3, label='Taylor Series (degree 5)')
plt.title('ln(x) at x=1')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x_values, y_values4, label='tan(x)')
plt.plot(x_values, y_taylor4, label='Taylor Series (degree 5)')
plt.title('tan(x) at x=pi/4')
plt.legend()

plt.tight_layout()
plt.show()
```
